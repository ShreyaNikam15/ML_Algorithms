```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
df.head()
```

```python
# Importing the StandardScaler class from the sklearn.preprocessing module
from sklearn.preprocessing import StandardScaler

# Defining a list of feature names
features = ['sepal length', 'sepal width', 'petal length', 'petal width']

# Separating out the features from a DataFrame called 'df'
# ':' selects all rows, 'features' selects only the columns specified in the list
# '.values' converts the DataFrame to a NumPy array
x = df.loc[:, features].values

# Separating out the target variable from the DataFrame 'df'
# ':' selects all rows, '['target']' selects only the column named 'target'
# '.values' converts the DataFrame to a NumPy array
y = df.loc[:,['target']].values

# Standardizing the features by transforming them using StandardScaler
# 'fit_transform' fits the StandardScaler to the data and transforms it in one step
x = StandardScaler().fit_transform(x)
```

```python
# Importing the PCA (Principal Component Analysis) class from the sklearn.decomposition module
from sklearn.decomposition import PCA

# Creating a PCA object with 2 principal components
pca = PCA(n_components=2)

# Applying PCA to the standardized features 'x' and transforming them into the principal components
principalComponents = pca.fit_transform(x)

# Creating a DataFrame to store the principal components with appropriate column names
principalDf = pd.DataFrame(data=principalComponents,
                            columns=['principal component 1', 'principal component 2'])
```

```python
# Concatenating the principal components DataFrame 'principalDf' with the 'target' column from the original DataFrame 'df'
# 'axis = 1' specifies that concatenation should be done along the columns
finalDf = pd.concat([principalDf, df[['target']]], axis=1)
```

```python
# Importing the matplotlib.pyplot module and renaming it as plt for ease of use
import matplotlib.pyplot as plt

# Creating a new figure with a specific size
fig = plt.figure(figsize=(8, 8))

# Adding a subplot to the figure
ax = fig.add_subplot(1, 1, 1)

# Setting labels for the x and y axes
ax.set_xlabel('Principal Component 1', fontsize=15)
ax.set_ylabel('Principal Component 2', fontsize=15)

# Setting the title of the plot
ax.set_title('2 component PCA', fontsize=20)

# Defining the target categories
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

# Defining colors for each target category
colors = ['r', 'g', 'b']

# Looping through each target category and corresponding color
for target, color in zip(targets, colors):
    # Selecting indices where the target category matches in the 'finalDf' DataFrame
    indicesToKeep = finalDf['target'] == target
    # Scatter plotting the principal components for the selected target category
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1'],
               finalDf.loc[indicesToKeep, 'principal component 2'],
               c=color,  # Setting the color
               s=50)     # Setting the size of the markers

# Adding a legend to the plot with target categories
ax.legend(targets)

# Adding grid lines to the plot
ax.grid()
```
